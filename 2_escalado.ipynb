### 1. ¿Qué diferencias encuentras entre las estadísticas originales y las del escalado MinMax?


```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

iris = load_iris(as_frame=True)
data = iris.data

scaler = MinMaxScaler()
data_scaled = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
```

### 2. Crea una gráfica de la distribución de la variable 'sepal length (cm)' antes y después del escalado.

```python
sepal_original = data['sepal length (cm)']
sepal_scaled = data_scaled['sepal length (cm)']

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.hist(sepal_original, bins=20, alpha=0.5, label='Original')
plt.hist(sepal_scaled, bins=20, alpha=0.5, label='Escalado')

plt.legend()
plt.title('Sepal length antes vs después (MinMax)')
plt.show()
```

### 3. Crea una gráfica de dispersión entre 'sepal length (cm)' y 'sepal width (cm)' con los datos escalados por StandardScaler.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_std = scaler.fit_transform(data)

data_std = pd.DataFrame(data_std, columns=data.columns)

plt.figure(figsize=(8,5))

plt.scatter(data_std['sepal length (cm)'], data_std['sepal width (cm)'])

plt.title('Sepal length vs width (StandardScaler)')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.show()
```

### 4. ¿Qué técnica de escalado crees que es más adecuada para este conjunto de datos? ¿Por qué?

**Respuesta:**

La técnica más adecuada para este conjunto de datos es **StandardScaler**.

Esto se debe a que las variables del dataset (como sepal length y sepal width) tienen distribuciones aproximadamente normales y no están limitadas a un rango específico. StandardScaler transforma los datos para que tengan una media de 0 y una desviación estándar de 1, lo que permite mantener la forma de la distribución original.

Además, este tipo de escalado suele ser más adecuado para muchos modelos de machine learning, como regresión, SVM y KNN, ya que mejora su rendimiento al trabajar con variables en escalas comparables.

Por otro lado, MinMaxScaler es útil cuando se necesita un rango específico (como 0 a 1), pero es más sensible a valores atípicos.
